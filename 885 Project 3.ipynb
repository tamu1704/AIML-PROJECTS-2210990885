```python
#which Workclass getting the Highest Salary?
def salary_gem(sal):
    if sal=='<50k':
        return 0
    else:
        return 1
```

```python
gem['new_encoded_sal']=gem['income'].apply(salary_gem)
gem.head(10)
```

```python
gem['income'].replace('<50k', '>50k', inplace=True)
```

```python
gem['income'] = pd.to_numeric(gem['income'], errors='coerce')
m=gem.groupby('workclass')['income'].mean()
n=m.max()
o=m.max()

print(f"The workclass with the highest average salary is '{n}' with an average salary of ${o:.2f}")
```

```python
p = gem[gem['income'] == '>50k']

q =p[p['gender'] == 'Male'].shape[0]
r =p[p['gender'] == 'Female'].shape[0]

s = gem[gem['gender'] == 'Male'].shape[0]
t = gem[gem['gender'] == 'Female'].shape[0]

male_high_salary_proportion = q / s
female_high_salary_proportion = r / t

print(f"Proportion of males earning >$50k: {male_high_salary_proportion:.2f}")
print(f"Proportion of females earning >$50k: {female_high_salary_proportion:.2f}")

if male_high_salary_proportion > female_high_salary_proportion:
    print("Males have a better chance of earning >$50k.")
elif female_high_salary_proportion > male_high_salary_proportion:
    print("Females have a better chance of earning >$50k.")
else:
    print("Males and females have an equal chance of earning >$50k.")
```
